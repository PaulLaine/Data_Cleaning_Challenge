```python
import pandas as pd
```

```python
df = pd.read_csv('data_cleaning_challenge.csv')
df
```

We first remove inconsistent features

```python
df_drop = df.drop(["Unnamed: 9", "Unnamed: 10"], axis = 1)
df_drop
```

There is some space between each person so we can remove each NaN values in the "Row Type" feature

```python
df_nan = df_drop[df_drop["Row Type"].notna()]
df_nan
```

```python
group = []
counter = 0

df_group = df_nan.copy()

for i in df_group["Row Type"]:
    if "first name: Person" in i:
        counter += 1
    group.append(counter)

df_group["Group"] = group
df_group
```

We can remove each row where it's a duplicate with the header.

```python
df_drop_extra_row = df_group[df_group["Row Type"] != "Row Type"]
df_drop_extra_row
```

Now we can divide our dataset into two datasets where we will have the values and the non_values to finally join them.

We will apply some changes to the non_value dataset to make it more usefull and clear

```python
df_non_values = df_drop_extra_row[df_drop_extra_row["Row Type"].str.contains("first name")]
df_non_values.drop(columns=["Speed1", "Speed2", "Electricity", "Effort", "Weight", "Torque"], inplace=True)
df_non_values.rename(columns={"Row Type":"First name", "Iter Number": "Last name", "Power1":"Date"}, inplace=True)
df_non_values["First name"] = df_non_values["First name"].str[12:]
df_non_values["Last name"] = df_non_values["Last name"].str[11:]
df_non_values["Date"] = df_non_values["Date"].str[6:]

df_non_values
```

We then create our value dataset

```python
df_values = df_drop_extra_row[~df_drop_extra_row["Row Type"].str.contains("first name")]

df_values
```

We can now join our two groups thanks to the "Group" feature.

```python
df_merge = df_non_values.merge(df_values, how='inner', on='Group')
df_merge
```

We can now remove our "Group" feature and we are good !

```python
df_final = df_merge.drop(["Group"], axis = 1)
df_final
```
